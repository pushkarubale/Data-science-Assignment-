```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import association_rules,apriori
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load Dataset
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/pushkarubale/Data-science-Assignment-/refs/heads/main/Online%20retail.csv')
df
```

```python
df.describe()
```

```python
txt = 'burgers/meatballs/eggs'
type(txt)
txt.split('/')
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
txt_split('python,session,machine,learning')
```

```python
df.head(5)
```

```python
data = df['shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil'].apply(txt_split)
```

```python
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
#data.replace(True,1,inplace=True)
# data.replace(False,0,inplace=True)
```

```python
data.replace([True,False],[1,0], inplace=True)
```

```python
data
```

```python
data.shape
```

```python
scores = apriori(data,min_support = 0.2, use_colnames=True)
scores
```

```python
rules = association_rules(scores)
rules
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
